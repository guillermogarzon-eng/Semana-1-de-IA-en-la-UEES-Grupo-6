```python
#Convertir algunas columnas numéricas a arrays NumPy:
age = titanic['age'].to_numpy()
fare = titanic['fare'].to_numpy()
pclass = titanic['pclass'].to_numpy()
survived = titanic['survived'].to_numpy()
print(age)
print(fare)
print(pclass)
print(survived)
```

```python
#Reemplazar edades faltantes (NaN) por la media
age_filled = age.copy()
age_filled[np.isnan(age_filled)] = np.nanmean(age_filled)
print(age_filled, round (2))
```

```python
#Crear un array 2D (matriz) con [edad, clase, tarifa]
matrix = np.column_stack((age, pclass, fare))
print(matrix, round (2))
```

```python
#Calcular el máximo, mínimo y promedio del precio del billete
fare_min = np.nanmin(fare)
fare_max = np.nanmax(fare)
fare_mean = np.nanmean(fare)
print("valor min billete", "=", f"{fare_min:.2f}")
print("valor max billete", "=", f"{fare_max:.2f}")
print("valor prom billete", "=", f"{fare_mean:.2f}")
```
